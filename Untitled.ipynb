```python
# ============================================================
# End-to-end checkpoint selector + comparison visualizer
# for:
#   BOTH_DIR   = both-task model eval_results
#   ROBOT_DIR  = robot-only model eval_results
#   FORK_DIR   = forklift-only model eval_results
#
# What it does:
# 1) scans the 3 eval_results folders
# 2) keeps checkpoint-*.json and optionally final.json
# 3) selects the best checkpoint separately for:
#       - forklift-only task
#       - robot-only task
#       - both-model on forklift task
#       - both-model on robot task
# 4) prepares comparison tables
# 5) renders fallback visuals
#
# If you already have plotting code and want EXACT same visuals,
# keep your old plotting section and use the dataframes produced
# by this script:
#   - forklift_overall_df
#   - forklift_field_acc_df
#   - forklift_field_f1_df
#   - robot_overall_df
#   - robot_field_acc_df
#   - robot_field_f1_df
# ============================================================

from __future__ import annotations

import json
import math
import re
from pathlib import Path
from typing import Any, Dict, List, Optional, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# ============================================================
# CONFIG
# ============================================================

BOTH_DIR = Path("runs/qwen35_9b_both_aug_2/eval_results")
ROBOT_DIR = Path("runs/qwen35_9b_robot_aug_all/eval_results")
FORK_DIR = Path("runs/qwen35_9b_v3/eval_results")

OUTPUT_DIR = Path("best_checkpoint_comparison_outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

INCLUDE_FINAL_JSON = True

# selection metric priority
# 1) exact_match_accuracy
# 2) relevant_field_accuracy
# 3) fewer parse_failures
# 4) fewer ground_truth_parse_failures
# 5) prefer final.json only if still tied
# 6) later checkpoint step
TASK_FIELDS = {
    "forklift": ["hazard_label", "hazard_present", "zone_relation", "object_state", "object_direction"],
    "robot": ["hazard_label", "hazard_present", "zone_relation", "object_state"],
}


# ============================================================
# FILE DISCOVERY / LOADING
# ============================================================

def is_candidate_eval_file(path: Path, include_final: bool = True) -> bool:
    if not path.is_file():
        return False
    if path.suffix.lower() != ".json":
        return False

    name = path.name.lower()

    if re.fullmatch(r"checkpoint-\d+\.json", name):
        return True

    if include_final and name == "final.json":
        return True

    return False


def checkpoint_step_from_name(filename: str) -> int:
    if filename == "final.json":
        return 10**12
    m = re.fullmatch(r"checkpoint-(\d+)\.json", filename)
    if m:
        return int(m.group(1))
    return -1


def load_json(path: Path) -> Dict[str, Any]:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


# ============================================================
# METRIC EXTRACTION
# ============================================================

def get_metrics_root(payload: Dict[str, Any]) -> Dict[str, Any]:
    return payload.get("metrics", {})


def get_task_block(payload: Dict[str, Any], task: str) -> Optional[Dict[str, Any]]:
    """
    Prefer metrics.by_task[task].
    If absent, fall back to combined only if that file is effectively single-task.
    """
    metrics = get_metrics_root(payload)

    by_task = metrics.get("by_task", {})
    if isinstance(by_task, dict) and task in by_task and by_task[task]:
        return by_task[task]

    combined = metrics.get("combined", {})
    task_counts = combined.get("task_counts", {})
    if isinstance(task_counts, dict) and task_counts:
        nonzero_tasks = {k for k, v in task_counts.items() if (v or 0) > 0}
        if nonzero_tasks == {task}:
            return combined

    return None


def safe_float(x: Any, default: float = float("-inf")) -> float:
    try:
        if x is None:
            return default
        return float(x)
    except Exception:
        return default


def selection_key(payload: Dict[str, Any], task: str, file_path: Path) -> Tuple[float, float, float, float, int, int]:
    block = get_task_block(payload, task)
    if block is None:
        return (float("-inf"), float("-inf"), float("-inf"), float("-inf"), 0, -1)

    exact = safe_float(block.get("exact_match_accuracy"))
    relevant = safe_float(block.get("relevant_field_accuracy"))

    parse_failures = block.get("parse_failures", 10**9)
    gt_parse_failures = block.get("ground_truth_parse_failures", 10**9)

    # fewer failures is better, so negate
    parse_failures_score = -float(parse_failures)
    gt_parse_failures_score = -float(gt_parse_failures)

    is_final = 1 if file_path.name == "final.json" else 0
    step = checkpoint_step_from_name(file_path.name)

    return (
        exact,
        relevant,
        parse_failures_score,
        gt_parse_failures_score,
        is_final,
        step,
    )


def get_per_field_accuracy(block: Dict[str, Any], field: str) -> Optional[float]:
    return block.get("per_field_accuracy", {}).get(field)


def get_per_field_metric(block: Dict[str, Any], field: str, metric_name: str) -> Optional[float]:
    field_block = block.get("per_field_metrics", {}).get(field, {})
    if not isinstance(field_block, dict):
        return None
    return field_block.get(metric_name)


def get_hazard_binary_metric(block: Dict[str, Any], metric_name: str) -> Optional[float]:
    return block.get("hazard_present_binary", {}).get(metric_name)


def get_hazard_binary_fp(block: Dict[str, Any]) -> Optional[int]:
    fp = block.get("hazard_present_binary", {}).get("fp")
    if fp is None:
        return None
    return int(fp)


def get_hazard_binary_tn(block: Dict[str, Any]) -> Optional[int]:
    tn = block.get("hazard_present_binary", {}).get("tn")
    if tn is None:
        return None
    return int(tn)


def false_positive_rate(block: Dict[str, Any]) -> Optional[float]:
    fp = get_hazard_binary_fp(block)
    tn = get_hazard_binary_tn(block)
    if fp is None or tn is None:
        return None
    denom = fp + tn
    if denom == 0:
        return None
    return fp / denom


# ============================================================
# CANDIDATE COLLECTION
# ============================================================

def collect_candidates(folder: Path, task: str, experiment_label: str) -> pd.DataFrame:
    if not folder.exists():
        raise FileNotFoundError(f"Folder does not exist: {folder}")

    rows = []

    for path in sorted(folder.iterdir()):
        if not is_candidate_eval_file(path, include_final=INCLUDE_FINAL_JSON):
            continue

        try:
            payload = load_json(path)
        except Exception as e:
            print(f"[WARN] Failed to load {path}: {e}")
            continue

        task_block = get_task_block(payload, task)
        if task_block is None:
            continue

        skey = selection_key(payload, task, path)

        row = {
            "experiment_label": experiment_label,
            "task": task,
            "file_path": str(path),
            "file_name": path.name,
            "checkpoint_step": checkpoint_step_from_name(path.name),
            "is_final": path.name == "final.json",
            "selection_exact_match_accuracy": skey[0],
            "selection_relevant_field_accuracy": skey[1],
            "selection_parse_failures_score": skey[2],
            "selection_gt_parse_failures_score": skey[3],
            "selection_is_final_score": skey[4],
            "selection_step_score": skey[5],
            "payload": payload,
        }

        rows.append(row)

    if not rows:
        raise RuntimeError(f"No usable candidate eval files found in {folder} for task='{task}'")

    df = pd.DataFrame(rows)

    df = df.sort_values(
        by=[
            "selection_exact_match_accuracy",
            "selection_relevant_field_accuracy",
            "selection_parse_failures_score",
            "selection_gt_parse_failures_score",
            "selection_is_final_score",
            "selection_step_score",
        ],
        ascending=[False, False, False, False, False, False],
    ).reset_index(drop=True)

    return df


def select_best_candidate(folder: Path, task: str, experiment_label: str) -> Dict[str, Any]:
    df = collect_candidates(folder, task, experiment_label)
    return df.iloc[0].to_dict()


# ============================================================
# SUMMARY BUILDERS
# ============================================================

def build_overall_summary(payload: Dict[str, Any], task: str, model_label: str, source_file: str) -> Dict[str, Any]:
    block = get_task_block(payload, task)
    if block is None:
        raise ValueError(f"Task block '{task}' not found in payload")

    return {
        "model": model_label,
        "task": task,
        "source_file": source_file,
        "total_samples": block.get("total_samples"),
        "exact_match_accuracy": block.get("exact_match_accuracy"),
        "relevant_field_accuracy": block.get("relevant_field_accuracy"),
        "parse_failures": block.get("parse_failures"),
        "ground_truth_parse_failures": block.get("ground_truth_parse_failures"),
        "hazard_precision": get_hazard_binary_metric(block, "precision"),
        "hazard_recall": get_hazard_binary_metric(block, "recall"),
        "hazard_f1": get_hazard_binary_metric(block, "f1"),
        "hazard_fp_count": get_hazard_binary_fp(block),
        "hazard_fp_rate": false_positive_rate(block),
    }


def build_field_summary(payload: Dict[str, Any], task: str, model_label: str, source_file: str) -> pd.DataFrame:
    block = get_task_block(payload, task)
    if block is None:
        raise ValueError(f"Task block '{task}' not found in payload")

    rows = []
    for field in TASK_FIELDS[task]:
        acc = get_per_field_accuracy(block, field)
        macro_f1 = get_per_field_metric(block, field, "macro_f1")
        weighted_f1 = get_per_field_metric(block, field, "weighted_f1")
        num_scored = get_per_field_metric(block, field, "num_scored_samples")

        if acc is None and macro_f1 is None and weighted_f1 is None:
            continue

        rows.append(
            {
                "model": model_label,
                "task": task,
                "source_file": source_file,
                "field": field,
                "accuracy": acc,
                "macro_f1": macro_f1,
                "weighted_f1": weighted_f1,
                "num_scored_samples": num_scored,
            }
        )

    return pd.DataFrame(rows)


def build_pair_tables(
    specific_best: Dict[str, Any],
    both_best: Dict[str, Any],
    task: str,
    specific_label: str,
    both_label: str = "Both-tasks",
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    specific_payload = specific_best["payload"]
    both_payload = both_best["payload"]

    overall_rows = [
        build_overall_summary(
            specific_payload,
            task=task,
            model_label=specific_label,
            source_file=specific_best["file_name"],
        ),
        build_overall_summary(
            both_payload,
            task=task,
            model_label=both_label,
            source_file=both_best["file_name"],
        ),
    ]
    overall_df = pd.DataFrame(overall_rows)

    field_df = pd.concat(
        [
            build_field_summary(
                specific_payload,
                task=task,
                model_label=specific_label,
                source_file=specific_best["file_name"],
            ),
            build_field_summary(
                both_payload,
                task=task,
                model_label=both_label,
                source_file=both_best["file_name"],
            ),
        ],
        ignore_index=True,
    )

    return overall_df, field_df


# ============================================================
# FALLBACK VISUALS
# ============================================================

def _annotate_bars(ax, rects, fmt="{:.3f}", rotation=0):
    for r in rects:
        h = r.get_height()
        if h is None or (isinstance(h, float) and np.isnan(h)):
            continue
        ax.annotate(
            fmt.format(h),
            xy=(r.get_x() + r.get_width() / 2, h),
            xytext=(0, 3),
            textcoords="offset points",
            ha="center",
            va="bottom",
            fontsize=9,
            rotation=rotation,
        )


def plot_overall_metrics(overall_df: pd.DataFrame, task: str, save_path: Optional[Path] = None):
    metrics = [
        "exact_match_accuracy",
        "relevant_field_accuracy",
        "hazard_precision",
        "hazard_recall",
        "hazard_f1",
    ]
    display_names = [
        "Exact Match",
        "Relevant Field Acc",
        "Hazard Precision",
        "Hazard Recall",
        "Hazard F1",
    ]

    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(metrics))
    n_models = len(overall_df)
    width = 0.8 / max(n_models, 1)

    for i, (_, row) in enumerate(overall_df.iterrows()):
        vals = [row[m] if row[m] is not None else np.nan for m in metrics]
        rects = ax.bar(x + (i - (n_models - 1) / 2) * width, vals, width, label=row["model"])
        _annotate_bars(ax, rects)

    ax.set_title(f"{task.capitalize()} - Best Checkpoint Comparison (Overall Metrics)")
    ax.set_xticks(x)
    ax.set_xticklabels(display_names, rotation=15)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel("Score")
    ax.legend()
    ax.grid(axis="y", linestyle="--", alpha=0.4)

    # add checkpoint names below title as a subtitle-like text
    subtitle = " | ".join([f'{r["model"]}: {r["source_file"]}' for _, r in overall_df.iterrows()])
    fig.text(0.5, 0.01, subtitle, ha="center", fontsize=9)

    fig.tight_layout(rect=[0, 0.04, 1, 1])

    if save_path:
        fig.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()


def plot_field_metric(field_df: pd.DataFrame, task: str, value_col: str, title_suffix: str, save_path: Optional[Path] = None):
    pivot = field_df.pivot(index="field", columns="model", values=value_col)

    fields_in_order = [f for f in TASK_FIELDS[task] if f in pivot.index]
    pivot = pivot.loc[fields_in_order]

    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(pivot.index))
    model_names = list(pivot.columns)
    n_models = len(model_names)
    width = 0.8 / max(n_models, 1)

    for i, model in enumerate(model_names):
        vals = pivot[model].values.astype(float)
        rects = ax.bar(x + (i - (n_models - 1) / 2) * width, vals, width, label=model)
        _annotate_bars(ax, rects)

    ax.set_title(f"{task.capitalize()} - Best Checkpoint Comparison ({title_suffix})")
    ax.set_xticks(x)
    ax.set_xticklabels(pivot.index, rotation=20)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel(value_col.replace("_", " ").title())
    ax.legend()
    ax.grid(axis="y", linestyle="--", alpha=0.4)

    subtitle_parts = []
    tmp = field_df[["model", "source_file"]].drop_duplicates()
    for _, r in tmp.iterrows():
        subtitle_parts.append(f'{r["model"]}: {r["source_file"]}')
    fig.text(0.5, 0.01, " | ".join(subtitle_parts), ha="center", fontsize=9)

    fig.tight_layout(rect=[0, 0.04, 1, 1])

    if save_path:
        fig.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()


def plot_false_positive_comparison(overall_df: pd.DataFrame, task: str, save_path: Optional[Path] = None):
    metrics = ["hazard_fp_count", "hazard_fp_rate"]
    display_names = ["Hazard FP Count", "Hazard FP Rate"]

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    for ax, metric, display_name in zip(axes, metrics, display_names):
        vals = overall_df[metric].astype(float).values
        rects = ax.bar(overall_df["model"], vals)
        if metric == "hazard_fp_rate":
            ax.set_ylim(0, 1.05)
        ax.set_title(display_name)
        ax.grid(axis="y", linestyle="--", alpha=0.4)
        _annotate_bars(ax, rects)

    fig.suptitle(f"{task.capitalize()} - Hazard False Positive Comparison")
    fig.tight_layout()

    if save_path:
        fig.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()


# ============================================================
# OPTIONAL: EXACT OLD VISUALS HOOK
# Paste your old plotting functions here if needed.
#
# Expected prepared inputs:
#   forklift_overall_df
#   forklift_field_acc_df
#   forklift_field_f1_df
#   robot_overall_df
#   robot_field_acc_df
#   robot_field_f1_df
# ============================================================

# def your_old_plotting_code(...):
#     pass


# ============================================================
# MAIN SELECTION
# ============================================================

# best for forklift-specific experiment
best_fork_specific = select_best_candidate(FORK_DIR, task="forklift", experiment_label="Forklift-only")

# best for robot-specific experiment
best_robot_specific = select_best_candidate(ROBOT_DIR, task="robot", experiment_label="Robot-only")

# best BOTH checkpoint for forklift task
best_both_on_forklift = select_best_candidate(BOTH_DIR, task="forklift", experiment_label="Both-tasks")

# best BOTH checkpoint for robot task
best_both_on_robot = select_best_candidate(BOTH_DIR, task="robot", experiment_label="Both-tasks")


# ============================================================
# PRINT SELECTED CHECKPOINTS
# ============================================================

def print_selected(best_row: Dict[str, Any], title: str):
    payload = best_row["payload"]
    block = get_task_block(payload, best_row["task"])
    print("=" * 90)
    print(title)
    print("-" * 90)
    print(f"Experiment     : {best_row['experiment_label']}")
    print(f"Task           : {best_row['task']}")
    print(f"Selected file  : {best_row['file_path']}")
    print(f"Exact Match    : {block.get('exact_match_accuracy')}")
    print(f"Relevant Acc   : {block.get('relevant_field_accuracy')}")
    print(f"Parse Failures : {block.get('parse_failures')}")
    print(f"GT Parse Fail  : {block.get('ground_truth_parse_failures')}")


print_selected(best_fork_specific, "BEST FORKLIFT-SPECIFIC CHECKPOINT")
print_selected(best_both_on_forklift, "BEST BOTH-MODEL CHECKPOINT ON FORKLIFT TASK")
print_selected(best_robot_specific, "BEST ROBOT-SPECIFIC CHECKPOINT")
print_selected(best_both_on_robot, "BEST BOTH-MODEL CHECKPOINT ON ROBOT TASK")


# ============================================================
# BUILD COMPARISON TABLES
# ============================================================

forklift_overall_df, forklift_field_df = build_pair_tables(
    specific_best=best_fork_specific,
    both_best=best_both_on_forklift,
    task="forklift",
    specific_label="Forklift-only",
    both_label="Both-tasks",
)

robot_overall_df, robot_field_df = build_pair_tables(
    specific_best=best_robot_specific,
    both_best=best_both_on_robot,
    task="robot",
    specific_label="Robot-only",
    both_label="Both-tasks",
)

forklift_field_acc_df = forklift_field_df[["model", "task", "source_file", "field", "accuracy"]].copy()
forklift_field_f1_df = forklift_field_df[["model", "task", "source_file", "field", "macro_f1"]].copy()

robot_field_acc_df = robot_field_df[["model", "task", "source_file", "field", "accuracy"]].copy()
robot_field_f1_df = robot_field_df[["model", "task", "source_file", "field", "macro_f1"]].copy()


# ============================================================
# SAVE TABLES
# ============================================================

forklift_overall_df.to_csv(OUTPUT_DIR / "forklift_overall_comparison.csv", index=False)
forklift_field_df.to_csv(OUTPUT_DIR / "forklift_field_comparison.csv", index=False)

robot_overall_df.to_csv(OUTPUT_DIR / "robot_overall_comparison.csv", index=False)
robot_field_df.to_csv(OUTPUT_DIR / "robot_field_comparison.csv", index=False)

selection_summary_rows = [
    {
        "comparison": "forklift_specific_best",
        "experiment_label": best_fork_specific["experiment_label"],
        "task": best_fork_specific["task"],
        "file_name": best_fork_specific["file_name"],
        "file_path": best_fork_specific["file_path"],
        "exact_match_accuracy": get_task_block(best_fork_specific["payload"], "forklift").get("exact_match_accuracy"),
        "relevant_field_accuracy": get_task_block(best_fork_specific["payload"], "forklift").get("relevant_field_accuracy"),
    },
    {
        "comparison": "both_best_on_forklift",
        "experiment_label": best_both_on_forklift["experiment_label"],
        "task": best_both_on_forklift["task"],
        "file_name": best_both_on_forklift["file_name"],
        "file_path": best_both_on_forklift["file_path"],
        "exact_match_accuracy": get_task_block(best_both_on_forklift["payload"], "forklift").get("exact_match_accuracy"),
        "relevant_field_accuracy": get_task_block(best_both_on_forklift["payload"], "forklift").get("relevant_field_accuracy"),
    },
    {
        "comparison": "robot_specific_best",
        "experiment_label": best_robot_specific["experiment_label"],
        "task": best_robot_specific["task"],
        "file_name": best_robot_specific["file_name"],
        "file_path": best_robot_specific["file_path"],
        "exact_match_accuracy": get_task_block(best_robot_specific["payload"], "robot").get("exact_match_accuracy"),
        "relevant_field_accuracy": get_task_block(best_robot_specific["payload"], "robot").get("relevant_field_accuracy"),
    },
    {
        "comparison": "both_best_on_robot",
        "experiment_label": best_both_on_robot["experiment_label"],
        "task": best_both_on_robot["task"],
        "file_name": best_both_on_robot["file_name"],
        "file_path": best_both_on_robot["file_path"],
        "exact_match_accuracy": get_task_block(best_both_on_robot["payload"], "robot").get("exact_match_accuracy"),
        "relevant_field_accuracy": get_task_block(best_both_on_robot["payload"], "robot").get("relevant_field_accuracy"),
    },
]

selection_summary_df = pd.DataFrame(selection_summary_rows)
selection_summary_df.to_csv(OUTPUT_DIR / "selected_checkpoints_summary.csv", index=False)

print("\nSaved comparison tables to:")
print(f"  {OUTPUT_DIR.resolve()}")


# ============================================================
# RENDER FALLBACK VISUALS
# If you want your exact old visuals, replace this section with
# your old plotting block and feed it the prepared dataframes.
# ============================================================

plot_overall_metrics(
    forklift_overall_df,
    task="forklift",
    save_path=OUTPUT_DIR / "forklift_overall_metrics.png",
)

plot_field_metric(
    forklift_field_df,
    task="forklift",
    value_col="accuracy",
    title_suffix="Per-field Accuracy",
    save_path=OUTPUT_DIR / "forklift_field_accuracy.png",
)

plot_field_metric(
    forklift_field_df,
    task="forklift",
    value_col="macro_f1",
    title_suffix="Per-field Macro F1",
    save_path=OUTPUT_DIR / "forklift_field_macro_f1.png",
)

plot_false_positive_comparison(
    forklift_overall_df,
    task="forklift",
    save_path=OUTPUT_DIR / "forklift_false_positive_comparison.png",
)

plot_overall_metrics(
    robot_overall_df,
    task="robot",
    save_path=OUTPUT_DIR / "robot_overall_metrics.png",
)

plot_field_metric(
    robot_field_df,
    task="robot",
    value_col="accuracy",
    title_suffix="Per-field Accuracy",
    save_path=OUTPUT_DIR / "robot_field_accuracy.png",
)

plot_field_metric(
    robot_field_df,
    task="robot",
    value_col="macro_f1",
    title_suffix="Per-field Macro F1",
    save_path=OUTPUT_DIR / "robot_field_macro_f1.png",
)

plot_false_positive_comparison(
    robot_overall_df,
    task="robot",
    save_path=OUTPUT_DIR / "robot_false_positive_comparison.png",
)


# ============================================================
# OPTIONAL: SHOW TABLES IN NOTEBOOK
# ============================================================

print("\nForklift overall comparison:")
print(forklift_overall_df.to_string(index=False))

print("\nForklift field comparison:")
print(forklift_field_df.to_string(index=False))

print("\nRobot overall comparison:")
print(robot_overall_df.to_string(index=False))

print("\nRobot field comparison:")
print(robot_field_df.to_string(index=False))
```

```python
import seaborn as sns
import numpy as np
import pandas as pd

FIELD_DISPLAY_MAP = {
    "hazard_label": "Hazard label",
    "hazard_present": "Hazard present",
    "zone_relation": "Zone relation",
    "object_state": "Object state",
    "object_direction": "Object direction",
}

def _get_task_block(payload, task):
    return payload["metrics"]["by_task"][task]

def _get_field_fp_total(task_block, field):
    """
    Sum FP across all labels for a field.
    This gives the field-level false-positive count used for improvement:
    FP improvement = single-task FP - both-task FP
    """
    field_block = task_block.get("per_field_metrics", {}).get(field, {})
    labels_block = field_block.get("labels", {})
    return sum(label_stats.get("fp", 0) for label_stats in labels_block.values())

def build_field_delta_df(single_payload, both_payload, task, single_label="Single-task", both_label="Both"):
    single_block = _get_task_block(single_payload, task)
    both_block = _get_task_block(both_payload, task)

    rows = []
    for field in TASK_FIELDS[task]:
        single_acc = single_block.get("per_field_accuracy", {}).get(field)
        both_acc = both_block.get("per_field_accuracy", {}).get(field)

        single_f1 = (
            single_block.get("per_field_metrics", {})
            .get(field, {})
            .get("macro_f1")
        )
        both_f1 = (
            both_block.get("per_field_metrics", {})
            .get(field, {})
            .get("macro_f1")
        )

        single_fp = _get_field_fp_total(single_block, field)
        both_fp = _get_field_fp_total(both_block, field)

        # flipped here
        acc_delta = None if single_acc is None or both_acc is None else single_acc - both_acc
        f1_delta = None if single_f1 is None or both_f1 is None else single_f1 - both_f1
        fp_improvement = both_fp - single_fp 

        rows.append({
            "field": field,
            "field_display": FIELD_DISPLAY_MAP.get(field, field),
            "Accuracy Δ": acc_delta,
            "Macro F1 Δ": f1_delta,
            "FP improvement": fp_improvement,
            f"{single_label} FP": single_fp,
            f"{both_label} FP": both_fp,
        })

    return pd.DataFrame(rows)

# rebuild
robot_field_delta = build_field_delta_df(
    single_payload=best_robot_specific["payload"],
    both_payload=best_both_on_robot["payload"],
    task="robot",
)

forklift_field_delta = build_field_delta_df(
    single_payload=best_fork_specific["payload"],
    both_payload=best_both_on_forklift["payload"],
    task="forklift",
)

robot_field_delta = robot_field_delta[
    ["field_display", "Accuracy Δ", "Macro F1 Δ", "FP improvement"]
].copy()

forklift_field_delta = forklift_field_delta[
    ["field_display", "Accuracy Δ", "Macro F1 Δ", "FP improvement"]
].copy()

display(robot_field_delta)
display(forklift_field_delta)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_field_delta_heatmaps():
    sns.set_theme(style="white", font_scale=1.05)

    fig, axes = plt.subplots(
        1, 2,
        figsize=(13.5, 4.8),
        gridspec_kw={"wspace": 0.35}
    )

    heatmaps = [
        ("Robot", robot_field_delta, axes[0]),
        ("Forklift", forklift_field_delta, axes[1]),
    ]

    metric_cols = ["Accuracy Δ", "Macro F1 Δ", "FP improvement"]

    for title, df, ax in heatmaps:
        plot_df = df.set_index("field_display")[metric_cols].astype(float)

        vmax = np.nanmax(np.abs(plot_df.to_numpy()))
        vmax = max(vmax, 0.05)

        sns.heatmap(
            plot_df,
            annot=True,
            fmt=".3f",
            cmap="RdYlGn",
            center=0,
            vmin=-vmax,
            vmax=vmax,
            linewidths=0.6,
            linecolor="#f2f2f2",
            cbar=True,
            cbar_kws={"shrink": 0.82, "pad": 0.02},
            annot_kws={"size": 18},
            ax=ax
        )

        ax.set_title(
            f"{title} — Improvement",
            fontsize=15,
            fontweight="semibold",
            pad=8
        )
        ax.set_xlabel("")
        ax.set_ylabel("")

        ax.tick_params(axis="x", labelrotation=90, labelsize=11)
        ax.tick_params(axis="y", labelrotation=0, labelsize=12)

        for spine in ax.spines.values():
            spine.set_visible(False)

    plt.suptitle(
        "Field-level deltas\n(Accuracy Δ and F1 Δ: Single - Both, FP improvement: Both - Single)",
        fontsize=18,
        fontweight="bold",
        y=1.02
    )

    plt.subplots_adjust(left=0.08, right=0.98, top=0.78, bottom=0.16, wspace=0.35)
    plt.show()

plot_field_delta_heatmaps()
```
